```python
city_list_unique=set(city_list_original)
len(city_list_unique)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

#Build Query url
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
#Set up list to hold response info
lat=[]
long=[]
temp_max=[]
hum=[]
cld=[]
wind=[]
cit=[]
country=[]
date=[]
```

```python
nums= 1
print('Beginning Data retrieval')
print('------------------------')
for city in city_list_unique:
   response_json=requests.get(query_url+city).json()
   if str(response_json['cod'])=='404':
       print(f'Missing data city no. {nums} of {len(city_list_unique)}|Skipped {city}')
       nums=nums + 1
   if str(response_json['cod'])!='404':
       try:
           cit.append(city)
           lat.append(response_json['coord']['lat'])
           print(f'Retrieve data city no. {nums} of {len(city_list_unique)}|Added {city}')
           nums=nums+1
       except:
           print('NA')
       try:
           long.append(response_json['coord']['lon'])
       except:
           print('NA')
       try:
           temp_max.append(response_json['main']['temp_max'])
       except:
           print('NA0')
       try:
           hum.append(response_json['main']['humidity'])
       except:
           print('NA1')
       try:
           cld.append(response_json['clouds']['all'])
       except:
           print('NA2')
       try:
           wind.append(response_json['wind']['speed'])
       except:
           print('NA3')
       try:
           country.append(response_json['sys']['country'])
       except:
           print('NA4')
       try:
           date.append(response_json['dt'])
       except:
           print('NA5')
print('----------------------')
print('End of Data retrieval')
```

```python
print(f'city  {len(cit)} latitude {len(lat)} long {len(long)} temp {len(temp_max)} humidity {len(hum)}')
```

```python
weather_data=pd.DataFrame()
weather_dict={
   'City': cit,
   'Latitude': lat,
   'Longitude': long,
   'Max Temperature': temp_max,
   'Humidity': hum,
   'Cloudiness': cld,
   'Wind Speed': wind,
   'City Country': country,
   'City Date': date
}
weather_data=pd.DataFrame(weather_dict)
weather_data
```
